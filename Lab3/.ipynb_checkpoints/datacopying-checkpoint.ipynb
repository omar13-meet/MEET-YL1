```python
import csv
f = open('imawesome.csv','rb')
reader = csv.reader(f) 
index = 0
tar = []
nicotine = []
for row in reader:
    a = row[0].replace("\t"," ") 
    a = row[0].split()
    b = (map(float,a[1:]))
    if index != 0:
        tar.append(b[0])
        nicotine.append(b[1])
    index += 1
print tar
print nicotine
    
f.close() 
    
```

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('nicotine and tar retionship')
ax.set_xlabel('nicotine')
ax.set_ylabel('tar')

scatter(tar,nicotine)
```
